```python
from sklearn import datasets
import numpy as np
import sklearn.metrics as sm
import pandas as pd
from sklearn.cluster import KMeans
from matplotlib import pyplot as plt
```

```python
iris = datasets.load_iris()
```

```python
iris.target_names
```

```python
iris.feature_names
```

```python
iris.data
```

```python
x = pd.DataFrame(iris.data)
x.columns = ['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']
y = pd.DataFrame(iris.target)
y.columns = ['Targets']
```

```python
x
```

```python
y
```

```python
# Set the size of the plot
plt.figure(figsize=(14,7))


# Create a colormap
colormap = np.array(['red', 'lime', 'black'])

# Plot Sepal
plt.subplot(1, 2, 1)
plt.scatter(x.sepal length (cm),x.sepal width (cm),c=colormap[y.Targets], s=40)
plt.title('Sepal')

# Plot Petal
plt.subplot(1, 2, 2)
plt.scatter(x.petal length (cm), x.petal width (cm), c=colormap[y.Targets], s=40)
plt.title('Petal');
```

```python
kmeans = KMeans(n_clusters=3).fit(x)
centroids = kmeans.cluster_centers_
print(centroids)
```

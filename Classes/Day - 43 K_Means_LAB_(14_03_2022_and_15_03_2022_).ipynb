# Clustering using K-Means

Clustering is the task of grouping together a set of objects in a way that objects in the same cluster are more similar to each other than to objects in other clusters. Every cluster will have a centroid which is defined as the center of the cluster.

# Steps in K-Means algorithm

1. First, we initialize k points((clusters), called means, randomly.
2. We categorize each item to its closest mean and we update the meanâ€™s coordinates, which are the averages of the items categorized in that mean so far.
3. We repeat the process for a given number of iterations and at the end, we have our clusters.

```python
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.cluster import KMeans
import sklearn.metrics as sm

import pandas as pd
import numpy as np
```

```python
Data = {'x': [1,1,1,2,1,2,1,2],
        'y': [4,2,4,1,1,4,1,1],
        'z': [1,2,2,2,1,2,2,1]
       }
  
df = pd.DataFrame(Data,columns=['x','y','z'])
print(df)
```

```python
kmeans = KMeans(n_clusters=2).fit(df)
centroids = kmeans.cluster_centers_
print(centroids)
```

```python
fig = plt.figure(figsize=(12, 15))
a = fig.add_subplot(projection='3d')
a.scatter(df['x'], df['y'], df['z'],c= kmeans.labels_.astype(float), s=100, alpha=0.5)
plt.xlabel('x-Axis')
plt.ylabel('y-Axis')
a.set_zlabel('Z-Axis')
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=50)
```

```python
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.cluster import KMeans
import sklearn.metrics as sm

import pandas as pd
import numpy as np
```

```python
Data = {'x': [25,34,22,27,33,33,31,22,35,34,67,54,57,43,50,57,59,52,65,47,49,48,35,33,44,45,38,43,51,46],
        'y': [79,51,53,78,59,74,73,57,69,75,51,32,40,47,53,36,35,58,59,50,25,20,14,12,20,5,29,27,8,7]
       }
  
df = pd.DataFrame(Data,columns=['x','y'])
print(df)
```

```python
kmeans = KMeans(n_clusters=3).fit(df)
centroids = kmeans.cluster_centers_
print(centroids)
```

```python
plt.scatter(df['x'], df['y'], c= kmeans.labels_.astype(float), s=20, alpha=.75)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=50)
plt.show()
```

```python
kmeans = KMeans(n_clusters=4).fit(df)
centroids = kmeans.cluster_centers_
print(centroids)
```

```python
plt.scatter(df['x'], df['y'], c= kmeans.labels_.astype(float), s=50, alpha=0.5)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=50)
plt.show()
```

# Using IRIS dataset

```python
iris = datasets.load_iris()

iris.data
iris.feature_names
iris.target
iris.target_names
```

```python
iris.data
```

```python
iris.feature_names
```

Store the inputs as a Pandas Dataframe and set the column names

```python
x = pd.DataFrame(iris.data)
x.columns = ['Sepal_Length','Sepal_Width','Petal_Length','Petal_Width']

y = pd.DataFrame(iris.target)
y.columns = ['Targets']
```

```python
x
```

```python
y
```

# Visualising the data

```python
# Set the size of the plot
plt.figure(figsize=(20,6))

# Create a colormap
colormap = np.array(['red', 'lime', 'black'])

# Plot Sepal
plt.subplot(1, 2, 1)
plt.scatter(x.Sepal_Length, x.Sepal_Width, c=colormap[y.Targets], s=40)
plt.title('Sepal')

# Plot Petal
plt.subplot(1, 2, 2)
plt.scatter(x.Petal_Length, x.Petal_Width, c=colormap[y.Targets], s=40)
plt.title('Petal');
```

# Building the K Means Model

```python
model = KMeans(n_clusters=3,)
model.fit(x)

model.labels_
model.cluster_centers_
```

Let's plot the actual classes against the predicted classes from the K Means model.

```python
model.labels_
```

```python
# Set the size of the plot
plt.figure(figsize=(14,7))

# Create a colormap
colormap = np.array(['red', 'lime', 'black'])

# Plot the Original Classifications
plt.subplot(1, 2, 1)
plt.scatter(x.Petal_Length, x.Petal_Width, c=colormap[y.Targets], s=40)
plt.title('Real Classification')

# Plot the Models Classifications
plt.subplot(1, 2, 2)
plt.scatter(x.Petal_Length, x.Petal_Width, c=colormap[model.labels_], s=40)

plt.title('K Mean Clustering');
```

We are going to change the class labels

```python
# The fix, we convert all the 1s to 0s and 0s to 1s.
predY = np.choose(model.labels_, [1, 0, 2]).astype(np.int64)
print (model.labels_)
print (predY)
```

Now we can re plot the data as before but using predY instead of model.labels_

```python
# Set the size of the plot
plt.figure(figsize=(14,7))

# Create a colormap
colormap = np.array(['red', 'lime', 'black'])

# Plot Orginal
plt.subplot(1, 2, 1)
plt.scatter(x.Petal_Length, x.Petal_Width, c=colormap[y.Targets], s=40)
plt.title('Real Classification')

# Plot Predicted with corrected values
plt.subplot(1, 2, 2)
plt.scatter(x.Petal_Length, x.Petal_Width, c=colormap[predY], s=40)
plt.title('K Mean Clustering');
```

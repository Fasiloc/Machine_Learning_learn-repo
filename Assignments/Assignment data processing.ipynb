# Task 3

```python
# outlier treatment
```

```python
import numpy as np
```

```python
# Upper boundary values in Marketing_expense
```

```python
np.percentile(dataf.Marketing_expense,[99])
```

```python
np.percentile(dataf.Marketing_expense,[99])[0]
```

```python
uv1=np.percentile(dataf.Marketing_expense,[99])[0]
```

```python
dataf[(dataf.Marketing_expense>uv1)]
```

```python
dataf.Marketing_expense[(dataf.Marketing_expense>3*uv1)]=3*uv1
```

```python
dataf[(dataf.Marketing_expense>uv1)]
```
